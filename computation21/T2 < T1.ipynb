```python
def takeIntersection(s):
    ret = []
    for i in range(len(s[0])):
        t = [set(x[i]) for x in s]
        r = set()
        for tt in t:
            r |= set(tt)
        r = tuple(r)
        ret.append(r)
        
    return ret
```

```python
transBackOrder = []
for c in candidateElements:
    torder = takeIntersection(c)
    transBackOrder.append(torder)
```

```python
def insertTheta(order):
    ret = []
    for i in range(len(order)+1):
        order1 = list(order).copy()
        order1[i:i] = [10]
        for j in range(i+1, len(order1)+1):
            order2  = order1.copy()
            order2[j:j] = [9]
            ret.append(order2)
    return ret
```

```python
def isContaining(poset,order):
    s1, s2, s3, s4 = set(), set(), set(), set()
    for i in order:
        if i in [9,10]:
            continue
        if order.index(i) < order.index(9):
            s1.add(i)
        if order.index(i) > order.index(9):
            s2.add(i)
        if order.index(i) < order.index(10):
            s3.add(i)
        if order.index(i) > order.index(10):
            s4.add(i)
    s = [s1, s2, s3, s4]
    #print(s)
    for i in range(len(poset)):
        if set(poset[i]).issubset(s[i]):
            continue
        else:
            return False
    return True
```

```python
def isContained(sub,sup):
    for i in range(len(sub)):
        if set(sub[i]).issuperset(set(sup[i])):
            continue
        else:
            return False
    return True

def identifyIndex8(l):
    #dl,tl,rl,s1l,s2l,s3l
    d8 = ((1,),(),(0,),(1,))
    t8 = ((),(6,),(2,),())
    r8 = ((3,),(7,),(),(3,))
    s18 = ((),(),(4,),())
    s28 = ((5,),(),(),(5,))
    s38 = ((),(8,),(),())
    
    indexList = [d8,t8,r8,s18,s28,s38]
    portrait=[]
    for region in indexList:
        if isContained(l,region):
            portrait.append(1)
        else:
            portrait.append(0)           
    return tuple(portrait)
```

```python
# generate the prec and desc for each 0,..., 7 polynomials
#partialOrderBase =[(4,0),(0,1),(1,6),(2,6),(1,7),(6,7),(5,7),(2,3),(3,5),(1,5),(2,7),(3,7),(4,7)] 
partialOrderBase = poset
prec = dict()
desc = dict()
for i,j in partialOrderBase:
    if i not in desc:
        desc[i] = []
    desc[i].append(j)
    
    if j not in prec:
        prec[j] = []
    prec[j].append(i)
    
def genPrecDict(i,prec):
    
    if i not in prec:
        return set()
    
    ret = set(prec[i])
    while True:
        preRet = ret.copy()
        for j in preRet:
            if j in prec:
                ret |= set(prec[j])
        
        if ret == preRet:
            return ret
        
def genDescDict(i,desc):
    
    if i not in desc:
        return set()
    
    ret = set(desc[i])
    while True:
        preRet = ret.copy()
        for j in preRet:
            if j in desc:
                ret |= set(desc[j])
        
        if ret == preRet:
            return ret

preDict = dict()
desDict = dict()

for i in range(9):
    preDict[i] = genPrecDict(i,prec)
    desDict[i] = genDescDict(i,desc)
```

```python
# simplify the defining inequalities
def getMaixmal(l,desDict):
    ret = []
    for i in l:
        intersect = set()
        if i in desDict:
            intersect = set(l)&set(desDict[i])
        if len(intersect) ==0:
            ret.append(i)    
    return ret

def getMinimal(l,desDict):
    ret = []
    for i in l:
        intersect = set()
        if i in preDict:
            intersect = set(l)&set(preDict[i])
        if len(intersect) ==0:
            ret.append(i)    
    return ret
```

```python
def removeRedundant(l):
    ret = []
    
    for i in range(len(l)):
        flag = True
        for j in range(len(l)):
            if i!=j and isContained(l[i],l[j]):
                flag = False
                break
        if flag:
            ret.append(l[i])
    return ret
```

```python
def generatePairConnectedComponent(indexSet,atomicComponents):
    fUnion = []
    for i in range(len(indexSet)):
        curr = indexSet[i]
        toRemove = []
        for j in range(len(fUnion)):
            currUnion = fUnion[j]
            for k in range(len(currUnion)):
                if isConnected(indexSet[i],currUnion[k],atomicComponents):
                    #print(indexSet[i],currUnion[k])
                    toRemove.append(j)
                    break
        newUnion = [curr]
        for j in range(len(toRemove)):
            index = toRemove[j]
            newUnion+=fUnion[index]
        updatedUnion = []
        for j in range(len(fUnion)):
            if j not in toRemove:
                updatedUnion.append(fUnion[j])
        updatedUnion.append(newUnion)
        fUnion = updatedUnion
           
    return fUnion

def isOrderContained(order, po):
    for i in po[0]:
        if order.index(i) > order.index(9):
            return False
    
    for i in po[1]:
        if order.index(i) < order.index(9):
            return False
    
    for i in po[2]:
        if order.index(i) > order.index(10):
            return False
        
    for i in po[3]:
        if order.index(i) < order.index(10):
            return False
        
    return True

def isConnected(c1,c2, atomicComponents):
    intersection = takeIntersection([c1,c2])
    for atom in atomicComponents:
        if isOrderContained(atom,intersection):
            #print(atom)
            #print(intersection)
            return True
    return False
```

```python
def isAdjacentOrder(order1, order2):
    differ = []
    for i in range(len(order1)):
        if order1[i]!= order2[i]:
            differ.append(i)
            
    if len(differ)!= 2:
        return False
    
    if abs(differ[0] - differ[1]) == 1:
        return True
    else:
        return False

def isAdjacentOrders(orders1, orders2):
    for order1 in orders1:
        for order2 in orders2:
            if isAdjacentOrder(order1, order2):
                return True
    return False
```

```python
def isSample(r, po,theta1,theta2):
    for i in po[0]:
        if r[i] >= theta1:
            return False
    for i in po[1]:
        if r[i]  <= theta1:
            return False
        
    for i in po[2]:
        if r[i] >=theta2:
            return False
        
    for i in po[3]:
        if r[i]  <= theta2:
            return False
    return True

def generateCoveringIndex(parameter, CoveringSets):
    parameter[:2] = sorted(parameter[:2])[::-1]
    parameter[-2:] = sorted(parameter[-2:])[::-1]
    theta1, theta2, mu, delta, eta1, eta2, e, f, = parameter
    factor = (e-f)*mu
    #factor = 1

    dl2 = (e + f + f*eta2)/(e + f)
    du2 = (e + f + e*eta2)/(e + f)
    dl1 = (e + f + e*eta2)/(e + f)
    tl2 =(e*(2 + mu) + f*(2 + 2*delta + mu)*(1 + eta1 + eta2))/((e + f)*(2 + delta + mu))
    tu1 = (f*(2 + mu) + e*(2 + 2*delta + mu)*(1 + eta1 + eta2))/((e + f)*(2 + delta + mu))
    ru2 = (e*(2 + mu)*(1 + eta2) + f*(2 + 2*delta + mu)*(1 + eta1 + eta2))/((e + f)*(2 + delta + mu))
    rl1 = (e*(2 + mu)*(1 + eta2) + f*(2 + 2*delta + mu)*(1 + eta1 + eta2))/((e + f)*(2 + delta + mu))
    ru1 = (f*(2 + mu)*(1 + eta2) + e*(2 + 2*delta + mu)*(1 + eta1 + eta2))/((e + f)*(2 + delta + mu))
    sl12 = 1
    su22 = 1+eta2
    sl21 = 1+eta2
    su31 = 1+eta1+eta2

    scaledList1 = [dl1,tu1,rl1,ru1,sl21,su31]
    scaledList2 = [dl2,du2,tl2,ru2,sl12,su22]

    theta1List = [x/factor for x in scaledList1]
    theta2List = [x/factor for x in scaledList2] 
    
    r = theta2List + [theta1List[1], theta1List[3],theta1List[5]]
    
    
    for po in CoveringSets:
        if isSample(r,po,theta1,theta2):
            return (po,r)
        
    return None
```

```python
import json
connections = []
with open('../figure21.gv','r') as file:
    line = file.readline()
    while line:
        if '--' in line:
            connections.append(line)
        line = file.readline()
```

```python
connections = [x.strip().split('--') for x in connections]
```

```python
for i in range(len(connections)):
    data = connections[i]
    data = [x.replace('"','').strip() for x in data]
    connections[i] = data
```

```python
def str_key(x):
    if ':' in x:
        xs = x.split(':')
    else:
        xs  = [x, '0']
    
    return 10*int(xs[0]) + int(xs[1])
```

```python
keys = set()
for i in range(len(connections)):
    data = connections[i]
    keys.add(data[0])
    keys.add(data[1])
```

```python
keys = sorted(list(keys), key = str_key)
```

```python
adjacent_matrix = [[0 for _ in range(70)] for _ in range(70)]
for a,b in connections:
    i, j = keys.index(a), keys.index(b)
    adjacent_matrix[i][j], adjacent_matrix[j][i] = 1, 1
```
